Para cada ejercicio de la tarea cree una clase para que sea más sencillo distinguir entre ellos.

Cada clase hereda de la clase padre la cual recibe el nombre de Ejercicio. La clase padre cuenta con todas las operaciones lógicas que se necesitan para los ejercicios.

```python
class Ejercicio:
  def __init__(self):
    self.p = [True,False]
    self.q = [True,False]

  def negacion(self,p):
    return not p

  def conjuncion(self,p,q):
    return p and q

  def disyuncion(self,p,q):
    return p or q

  def implicacion(self,p,q):
    if p == True and q == False:
      return False
    else:
      return True

  def XOR(self,p,q):
    return (p and not q) or (not p and q)
```

En este primer ejercicio se crea una lista c, la cual va a almacenar los valores de verdad de a y b. Y también incluirá los valores de verdad al someter las proposiciones a la operacion lógica XOR.

```python
class Ejercicio1(Ejercicio):
  def __init__(self):
    super().__init__()
    self.c = []

  def resuelve_ejercicio(self):
    columnas = ['a','b','c']
    for a in self.p:
      for b in self.q:
        self.c.append([a,
                      b,
                  super().XOR(a,b)])
    self.c = pd.DataFrame(self.c, columns=columnas)
```

La clase del ejercicio 2 realiza todas las operaciones lógicas indicadas de dos formas:

- La primera es por medio de la creación de una función (En la clase padre) que realice la operación lógica de implicación.

- La segunda es por medio de equivalencia, la cual consigo por medio de la operacion lógica disyunción.

Aunque al final solo imprimo una de ellas, ya que solo se pide una tabla.

```python
class Ejercicio2(Ejercicio):
  def __init__(self):
    super().__init__()
    self.f1 = []
    self.f2 = []

  def resuelve_ejercicio(self):
    columnas = ['p','q','p->q','q->p','¬q->¬p','¬p->¬q']
    for p in self.p:
      for q in self.q:
        self.f1.append([p,
                        q,
                        super().implicacion(p,q),
                        super().implicacion(q,p),
                        super().implicacion(super().negacion(q),super().negacion(p)),
                        super().implicacion(super().negacion(p),super().negacion(q))])
        self.f2.append([p,
                        q,
                        super().disyuncion(super().negacion(p),q),
                        super().disyuncion(p, super().negacion(q)),
                        super().disyuncion(super().negacion(p),q),
                        super().disyuncion(p, super().negacion(q))])

    self.f1 = pd.DataFrame(self.f1, columns=columnas)
    self.f2 = pd.DataFrame(self.f2, columns=columnas)
    #print(self.f1)
    #print("\n")
    #print(self.f2)
```

```python
class Ejercicio3(Ejercicio):
  def __init__(self):
    super().__init__()
    self.resultado = []

  def calcula_resultado(self):
    nombre = ['p','q','¬p','¬q','pV¬q','(pV¬q)^¬p']
    for p in self.p:
      for q in self.q:
        self.resultado.append([p,
                               q,
                               super().negacion(p),
                               super().negacion(q),
                               super().disyuncion(p,super().negacion(q)),
                                super().conjuncion(super().disyuncion(p,super().negacion(q)),super().negacion(p))])


    self.resultado = pd.DataFrame(self.resultado, columns = nombre)
    #print(self.resultado)
```

#Ejercicio 4

La clase del ejercicio 4 realiza la tabla de verdad de las operaciones lógicas que éstas contenidas en la primera ley de DeMorgan que se muestra en el documento.

En las columnas se indica a que operación lógica corresponde o proposición, también contiene una columna que corresponde al resultado de comprobar si los resultados de las operaciones lógicas son equivalentes.

La tabla está hecha con las 4 combinaciones posibles de las proposiciones p y q.

```python
class Ejercicio4(Ejercicio):
  def __init__(self):
    super().__init__()
    self.prop1 = []
    self.prop2 = []

  def Q(self,p,q):
    return p == q

  def resuelve_ejercicio(self):
    self.resultado = []
    col1 = ['p','q','p^q','¬(p^q)']
    col2 = ['¬pV¬q','¬q','¬p']
    for p in self.p:
      for q in self.q:
        self.prop1.append([p,
                          q,
                          super().conjuncion(p,q),
                          super().negacion((super().conjuncion(p,q)))])
        self.prop2.append([super().negacion(p),
                          super().negacion(q),
                          super().disyuncion(super().negacion(p),super().negacion(q))])

    self.prop1 = pd.DataFrame(self.prop1, columns = col1)
    for i in range(len(self.prop2)):
      self.prop2[i] = list(reversed(self.prop2[i]))
    self.prop2 = pd.DataFrame(self.prop2, columns= col2)

    for i in range(len(self.prop1)):
      self.resultado.append(self.Q(self.prop1.iloc[i,-1],self.prop2.iloc[i,0]))

    self.resultado = pd.DataFrame(self.resultado, columns=['≡'])
    self.tabla = pd.concat([self.prop1,self.resultado,self.prop2], axis = 1)
```
